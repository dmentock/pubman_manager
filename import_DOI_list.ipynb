```python
from pubman_manager import DOIParser, PubmanBase, PUBLICATIONS_DIR
import datetime

import logging
pubman_api = PubmanBase()
doi_parser = DOIParser(pubman_api, logging_level = logging.DEBUG)


dois = ['10.1016/j.carbon.2024.119091', '10.1038/s41467-024-52123-w']

df_dois_overview = doi_parser.filter_dois(dois)
df_dois_overview
```

```python
dois_data = doi_parser.collect_data_for_dois(df_dois_overview)
pub_path = PUBLICATIONS_DIR / 'new' / f'DOI_list_{datetime.datetime.now().strftime("%d.%m.%Y")}.xlsx'
doi_parser.write_dois_data(pub_path, dois_data)
```
